## Gr√°ficas en 3 dimensiones

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D 
import numpy as np
```

$$z = x^2 - y^2$$

```python
def f(x,y):
    return x**2 - y**2

fig = plt.figure()
ax = fig.add_subplot(111, projection="3d")

x = np.linspace(-10,10,200)
y = np.linspace(-10,10,200)

x,y = np.meshgrid(x,y)

z = f(x,y)

ax.plot_surface(x,y,z)
plt.show()
```

$$z = x^2 + y^3x + y^5$$

```python
def f(x,y):
    return np.sin(np.sqrt(x**2 + y**2))

fig = plt.figure()
ax = fig.add_subplot(111, projection="3d")

x = np.linspace(-5,5,200)
y = np.linspace(-5,5,200)

x,y = np.meshgrid(x,y)

z = f(x,y)

ax.plot_surface(x,y,z)
plt.show()
```

```python
from matplotlib import cm

font_custom = { 'family': 'serif',
                'color':  'darkblue',
                'weight': 'bold',
                'size': 12}

def f(x,y):
    return x**2 - y**2

fig = plt.figure(figsize=(9,5))
ax = fig.add_subplot(111, projection="3d")

x = np.linspace(-10,10,200)
y = np.linspace(-10,10,200)

x,y = np.meshgrid(x,y)

z = f(x,y)

surf = ax.plot_surface(x,y,z,cmap=cm.inferno)

fig.colorbar(surf , shrink=0.8, aspect=10, label="color map")

ax.set_title("$z = x^2 - y^2$", fontdict=font_custom, loc="left")
ax.set_xlabel("$x$", fontdict=font_custom)
ax.set_ylabel("$y$", fontdict=font_custom)
ax.set_zlabel("$z$", fontdict=font_custom)
ax.tick_params(labelsize=8)
ax.set_xticks(np.linspace(-10,10,11))
ax.set_yticks(np.linspace(-10,10,11))
plt.savefig("3dimg01.pdf",dpi=300,bbox_inches="tight")
plt.show()
```

# Ecuaciones Parametricas

```python
def f(t):
    return [np.sin(t),np.cos(t),t]

t = np.linspace(0,50,400)

fig = plt.figure()
ax = fig.add_subplot(111, projection="3d")

x,y,z = f(t)

plt.plot(x,y,z)
```
