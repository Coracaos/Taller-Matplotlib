```python
import matplotlib.pyplot as plt
import numpy as np

def f(x):
    return np.abs(np.cos(x))

plt.style.use("ggplot")

fig = plt.figure()
ax = fig.add_subplot(111)

x = np.linspace(0,6*np.pi,100)
y = f(x)

ax.plot(x, y)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

def f(theta):
    return [np.sin(theta), np.cos(theta)] 

plt.style.use("ggplot")

fig = plt.figure()
ax = fig.add_subplot(111, aspect='equal')

theta = np.linspace(0, 2*np.pi, 200)
x,y = f(theta)

ax.plot(x, y)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

def f(x):
    return np.sqrt(1 - (np.abs(x) - 1)**2)

def g(x):
    return np.arccos(1-np.abs(x)) - np.pi

plt.style.use("ggplot")

fig = plt.figure()
ax = fig.add_subplot(111)

x = np.linspace(-2,2,1000)

ax.plot(x, f(x), color="red", linestyle="--", lw=3)
ax.plot(x, g(x), color="red", linestyle="--", lw=3)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

def f(x):
    return np.tan(x)

plt.style.use("ggplot")

fig = plt.figure()
ax = fig.add_subplot(111)

x = np.linspace(0,10,1001)
y = f(x)

y[y>20] = np.inf
y[y<-20] = -np.inf

ax.plot(x, y)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use("ggplot")

def f(x,p):
    return x**p
    
fig = plt.figure()
ax1 = fig.add_subplot(221)
ax2 = fig.add_subplot(222)
ax3 = fig.add_subplot(223)
ax4 = fig.add_subplot(224)

x = np.linspace(-5,5,100)

ax1.plot(x,f(x,1))
ax2.plot(x,f(x,2))
ax3.plot(x,f(x,3))
ax4.plot(x,f(x,4))

plt.savefig("img01.pdf", dpi=300)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use("ggplot")

def f(x,p):
    return x**p
    
fig , axs = plt.subplots(2,2)

axs = np.reshape(axs, axs.size)

for i in range(4): 
    axs[i].plot(x,f(x,i+1))
    
plt.subplots_adjust(hspace=0.30, wspace=0.30)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 4*np.pi, 101)

plt.plot(x, np.cos(x), label="$\cos(x)$")
plt.plot(x, np.sin(x), label="$\sin(x)$")

plt.title("Grafica de la función coseno y seno")
plt.xlabel("x")
plt.ylabel("y")
plt.legend(bbox_to_anchor=(1,1))

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure()
ax = fig.add_subplot(111)

x = np.linspace(0, 4*np.pi, 101)

ax.plot(x, np.cos(x), label="$\cos(x)$")
ax.plot(x, np.sin(x), label="$\sin(x)$")

ax.set_title("Grafica de la función coseno y seno")
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.legend(bbox_to_anchor=(1,1))
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

fig, axs = plt.subplots(2,1)
x = np.linspace(0, 6*np.pi, 101)

axs[0].plot(x,np.cos(x),label="$\cos(x)$")
axs[1].plot(x,np.sin(x),label="$\sin(x)$")

axs[0].legend(loc=2, bbox_to_anchor = (1,1))
axs[1].legend(loc=2, bbox_to_anchor = (1,1))

plt.suptitle("Grafica de coseno y seno")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import calendar as cal

plt.style.use("ggplot")

y1 = np.random.randint(1,20,12)
y2 = np.random.randint(1,20,12)

x = np.arange(1,13,1)

plt.plot(x, y1, 'bo--', x, y2, 'rD--')
plt.title("Temperatura registrada el año 2017", fontsize=15, color="#45426E")
plt.xticks(x, cal.month_abbr[1:13], rotation=20)
plt.legend(("Cuzco","Loreto"), loc=2, bbox_to_anchor=(1,1))
plt.ylim(np.min(np.concatenate((y1,y2))) - 3, np.max(np.concatenate((y1,y2))) + 3)
plt.xlabel("Meses")
plt.ylabel("Temperatura (C°)")
plt.savefig("clima.pdf", dpi=300, bbox_inches="tight")
plt.show()
```
